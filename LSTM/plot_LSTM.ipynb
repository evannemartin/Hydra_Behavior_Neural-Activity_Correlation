```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
totalloss = np.load("RestCoBenPas200/New/totalloss.npy")
time = np.load("RestCoBenPas200/New/timetrain.npy")
#time=np.arange(100)
plt.plot(time, totalloss)
plt.xlabel("Nb of epochs",fontsize=14)
plt.ylabel("Loss",fontsize=14)
plt.title('Loss depending on the nb of epochs', fontsize=18)
plt.ylim(0,0.7)
plt.show()
```

```python
time = np.load('RestCoBenPas200/New/timetest.npy')
test_output = np.load('RestCoBenPas200/New/expected.npy')
predicted = np.load('RestCoBenPas200/New/predicted.npy')

print(predicted)

for j in range(test_output.shape[0]) :
    plt.plot(time, test_output[j], label='expected')
    plt.plot(time, predicted[j], label='predicted')
    plt.title('Prediction for samples of size 200', fontsize=18)
    plt.xlabel("Nb of Frames", fontsize=14)
    plt.ylabel("0 : Rest*, 1 : Contraction, 2 : Left Bending", fontsize=12)
    plt.legend(loc='best')
    plt.show()
```
